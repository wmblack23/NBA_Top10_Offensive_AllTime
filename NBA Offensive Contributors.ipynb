# Who is *the best*?

>### Looking at the greatest offensive talents in NBA history

---

Answering the question: **"who is the greatest of all time"** is a timeless sports tradition.  Some debates are harder to settle than others.  But some questions, like **which player is the greatest offensive contributor in the history of the NBA**, have pretty clear metrics that we can point to when discussing.

- There are two MAIN statistics for offensive players: **points and assists**.  Some players are great scorers, while others are great passers, but the truly great offensive talents did both at the highest levels.

What we're looking at below are the **NBA's All-Time Leaders in Points + Assists** and diving into what the numbers tell us.

```python
import pandas as pd
from nba_api.stats.endpoints import commonallplayers
from nba_api.stats.endpoints import playercareerstats
from nba_api.stats.endpoints import alltimeleadersgrids
from matplotlib import pyplot as plt
import numpy as np
import random
```

## NBA Top 10 in Career Points + Assists
Adding total career points + total career assists and making a scatterplot of the Top 10 highest totals in league history.

```python
data = alltimeleadersgrids.AllTimeLeadersGrids()

df_pts = data.pts_leaders.get_data_frame().set_index('PLAYER_ID')
df_ast = data.ast_leaders.get_data_frame()

seasons = []
assists = []

for i in df_pts.index:
    
    stats_data = playercareerstats.PlayerCareerStats(i)
    df_stats = stats_data.season_totals_regular_season.get_data_frame()
    
    seasons.append(len(df_stats['SEASON_ID'].unique()))
    assists.append(sum(df_stats['AST']))
    
df_pts['Total'] = df_pts['PTS'] + assists
df_pts['Seasons Played'] = seasons

df_pts.set_index('PLAYER_NAME', inplace=True)

plt.scatter(df_pts['Seasons Played'], df_pts['Total'])
plt.xticks(np.arange(min(df_pts['Seasons Played']) - 1, max(df_pts['Seasons Played'])+2))
plt.title('Greatest Offensive Contributors - NBA History')
plt.xlabel('Seasons Played')
plt.ylabel('Total Career Points + Assists')

for i in df_pts.index:
    plt.annotate((i + ' - ' + df_pts.loc[i]['Total'].astype(str)),
                 (df_pts.loc[i]['Seasons Played'], df_pts.loc[i]['Total']),
                 (df_pts.loc[i]['Seasons Played'] + 0.4, df_pts.loc[i]['Total']),
                 arrowprops=dict(arrowstyle='-'))
```

### Points of interest
- Michael Jordan and Wilt Chamberlain played 14 and 15 years respectively, which is quite a bit less than everybody else in this Top 10. (LeBron and Carmelo have played the second least at 18 each, but they are still in the league.)
- By the time LeBron James retires, he will have created a pretty sizeable gap between himself and the rest of this group.
- That the gap between the top few guys and Moses Malone is almost 16,000 pts/ast shows just how incredible the top few are even when compared to other all time great offensive players.

## Average Points per Game + Assists per Game PER SEASON
Looking at our Top 5 contributors from above, this line graph shows how many points per game + assists per game they averaged for each season of their career.  These lines tell each players story: injuries, coming into the NBA at a young age, a swift fall-off at season x, etc.

```python
names = df_pts.sort_values(['Total'], ascending=False).index[0:5]

for l in names:
    
    player_data = commonallplayers.CommonAllPlayers()
    df_playerdata = player_data.common_all_players.get_data_frame().set_index('DISPLAY_FIRST_LAST')
    
    stats_data = playercareerstats.PlayerCareerStats(df_playerdata.loc[l]['PERSON_ID'])
    df_playerstats = stats_data.season_totals_regular_season.get_data_frame()
    
    points = df_playerstats.PTS/df_playerstats.GP
    assists = df_playerstats.AST/df_playerstats.GP
    total = points + assists
    total.index = np.arange(1, len(points) + 1)
    
    plt.title('Per Game/Per Season Numbers')
    plt.xlabel('Season of Career')
    plt.ylabel('Average Points + Assists Per Game')
    plt.plot(total, label = l)
    plt.xticks(np.arange(1, max(df_pts['Seasons Played']+1)))
    plt.legend()
```

### Points of interest
- Kobe had by far the toughest adjustment period into the NBA of any of these guys.  While him and LeBron both came into the league at 18, it took Kobe a number of years before he got to the level of the rest of the stars, whereas LeBron got there immediately.
- You can see why people say "Jordan had the greatest peak of all-time."  His run from seasons 3-9 are unparralelled.
- What's interesting to see is how Kobe and LeBron (before Kobe's achilles injury) were playing at SUCH a higher level by year 17+ than the other three.  By that time, Jordan was out of the league and Malone and Kareem were producing way lower than they once had.  There are two clear reasons for this, in my opinion:
    1. LeBron and Kobe both skipped high school.  So their "Year 17" came at around the age of 35, whereas Jordan, Kareem and Karl, who all played college, would have been closer to 38 years old in their 17th season.
    2. LeBron and Kobe both played the bulk of their careers from 2000 onwards.  Why is this significant?  The science of sports and nutrition and rehabilitation is light years ahead of where it was in the 80's and 90's.  We will probably start to see more and more players playing at these high levels deep into their career as time goes on.
- LeBron, in year 18, is producing more points and assists than Karl Malone, the 3rd greatest offensive contributor of all time, did during ANY season of his career.

## Career Points + Assists TOTALS by Year
Again using our Top 5, this graph charts each players total points + assists for each season of their career.

```python
player_data = commonallplayers.CommonAllPlayers()
df = player_data.common_all_players.get_data_frame().set_index('DISPLAY_FIRST_LAST')

for l in names:
    
    stats_data = playercareerstats.PlayerCareerStats(df.loc[l]['PERSON_ID'])
    df2 = stats_data.season_totals_regular_season.get_data_frame()
    
    total_stats = df2.PTS + df2.AST
    total_stats.index = np.arange(1, len(total_stats) + 1)
    
    for i in range(1, len(total_stats)+ 1):
        if i == 1:
            total_stats[i] = total_stats[i]
        else:
            total_stats[i] = total_stats[i] + total_stats[i - 1]
            
    plt.title('Career Totals')
    plt.xlabel('Season of Career')
    plt.ylabel('Total Career Points + Assists')
    plt.plot(total_stats, label = l)
    plt.xticks(np.arange(1, max(df_pts['Seasons Played'])+1))
    plt.legend()
```

### Points of interest
- Again, this graph will look pretty remarkable in just a few more seasons as LeBron continues to distance himself from the pack.
- You can see how far below the rest of the group Kobe started, yet because of his health and longevity was still able to catch and surpass Jordan.
- Malone is just incredibly consistent from year 2-13.
- Jordan and LeBron were neck and neck through 14 years of their NBA career.  Ultimately, what will set LeBron apart from Jordan is his health and longevity.

## Career Yearly Average
Lastly - while health and longevity are absolutely essential in determining "best ever" and should be looked at as a skill, it is hard not to recognize the advantages modern day players have in extending their careers.  So for that reason, I wanted to take a look at the **average total Points + Assists the Top 5 contributors had for their careers**.  That DataFrame is below.

```python
yearly_average = []
player_name = []

for l in names:
    
    player_data = commonallplayers.CommonAllPlayers()
    df = player_data.common_all_players.get_data_frame().set_index('DISPLAY_FIRST_LAST', drop=False)
    
    stats_data = playercareerstats.PlayerCareerStats(df.loc[l]['PERSON_ID'])
    df2 = stats_data.season_totals_regular_season.get_data_frame()

    yearly_average.append(int((sum(df2.PTS + df2.AST)) / (len(df2['SEASON_ID'].unique()))))
    player_name.append(df.loc[l]['DISPLAY_FIRST_LAST'])
    
    df_yearlyavg = pd.DataFrame()
    df_yearlyavg['Player Name'] = player_name
    df_yearlyavg['Yearly Points + Assists Average'] = yearly_average    
    
df_yearlyavg.sort_values(['Yearly Points + Assists Average'], ascending = False).set_index('Player Name')
```

### Points of interest
- There's a reason why LeBron and Jordan are usually discussed seperately.
- Kobe just couldn't overcome those first few years, statistically.
- LeBron being so close to Jordan while still playing into his late 30's AND having played in the NBA from 18-21 (which you would think would have brought his numbers down as he adjusted to the game) is incredible.

## Health Matters
Lastly - aside from a player having to be supremely talented to catch LeBron, HEALTH and AVAILABILITY are just as important of factors.  The chart below shows this by taking this same subset of young players as from above, but just charting their **games played per season**.

```python
names = (['LeBron James', 'Trae Young', 'Luka Doncic', 'Devin Booker', 'Zion Williamson',
          'Donovan Mitchell', 'Ben Simmons'])

player_data = commonallplayers.CommonAllPlayers()
df = player_data.common_all_players.get_data_frame().set_index('DISPLAY_FIRST_LAST')

for l in names:
    
    if l == 'LeBron James':
        
        stats_data = playercareerstats.PlayerCareerStats(df.loc[l]['PERSON_ID'])
        df2 = stats_data.season_totals_regular_season.get_data_frame()

        total_games = df2.GP
        total_games = total_games[0:7]
        total_games.index = np.arange(1, len(total_games) + 1)

        for i in range(1, len(total_games)+ 1):
            if i == 1:
                total_games[i] = total_games[i]
            else:
                total_games[i] = total_games[i] + total_games[i - 1]
                
    else:
        
    
        stats_data = playercareerstats.PlayerCareerStats(df.loc[l]['PERSON_ID'])
        df2 = stats_data.season_totals_regular_season.get_data_frame()

        total_games = df2.GP
        total_games = total_games[0:len(total_games)]
        total_games.index = np.arange(1, len(total_games) + 1)

        for i in range(1, len(total_games)+ 1):
            if i == 1:
                total_games[i] = total_games[i]
            else:
                total_games[i] = total_games[i] + total_games[i - 1]
            

    plt.plot(total_games, label = l)
    plt.xticks(np.arange(1, 8))
    plt.title('Total Games Played to Start NBA Career')
    plt.xlabel('Season of Career')
    plt.ylabel('Total Games Played')
    plt.legend()
```

### Points of interest
- As you can see, by Year 3 LeBron has distanced himself in terms of total games played from all of the young stars today.  Just another reason his record will be so challenging to ever catch.
